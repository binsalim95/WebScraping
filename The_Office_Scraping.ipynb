# The Office Scraper

In this notebook, I'll be scraping show info from the hit TV series, "The Office" off of its Wiki using the Beautiful Soup Package

```python
#Import
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
#Get Content
r = requests.get("https://en.wikipedia.org/wiki/The_Office_(American_TV_series)")
soup = bs(r.content)
content = soup.prettify()
```

We'll be getting the info from the side bar of "The Office", wiki.

```python
info_box = soup.find(class_='infobox vevent')
print(info_box.prettify())
```

```python
def get_content_value(row_data):
  if row.find("li") is not None:
    return [li.get_text(" ", strip = True).replace("\xa0", ' ') for li in row.find_all('li')]
  else:
    return row.find('td').get_text(" ", strip = True).replace("\xa0", ' ')
```

## Extracting
I'll be extracting all show info into a dictionary which we will then import into a Pandas Dataframe.

```python
show_info={}
X = info_box.find_all('tr')
for index, row in enumerate(X):
  if index==0:
    show_info['Title']=row.find('th').get_text()
  elif index == 1:
    continue
  else:
    x = row.find('th')
    y = row.find('td')
    if x is not None and y is not None:
      key = row.find('th').get_text()
      value = get_content_value(row.find('td'))
      show_info[key] = value

for key in show_info:
  print (f"{key}:{show_info[key]}")
```
